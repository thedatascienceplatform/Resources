```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Import data

```python
data = sns.load_dataset('tips')
data.head(5)
```

## Plotting

### Scatterplot

```python
sns.set_style(style='whitegrid')
fig, ax = plt.subplots(1,3, figsize = (16,5))
sns.scatterplot(x = 'tip', y = 'total_bill', data = data, ax = ax[0])
ax[0].set_title('Default Hue & Default Size')
sns.scatterplot(x = 'tip', y = 'total_bill', hue = 'sex', data = data, ax = ax[1])
ax[1].set_title('Custom Hue & Default Size')
sns.scatterplot(x = 'tip', y = 'total_bill', hue = 'sex', size = 'smoker', data = data, ax = ax[2])
ax[2].set_title('Custom Hue & Custom Size')
```

### Boxplot

```python
sns.set_style(style='white')
fig, ax = plt.subplots(1,3, figsize = (18,5))
sns.boxplot(x = 'sex', y = 'total_bill', hue = 'time', data = data, palette='rocket', ax = ax[0])
ax[0].set_title('Hue = time')
sns.boxplot(x = 'sex', y = 'total_bill', hue = 'smoker', data = data, palette='Spectral', ax = ax[1])
ax[1].set_title('Hue = smoker')
sns.boxplot(x = 'sex', y = 'total_bill', hue = 'day', data = data, palette='Set2', ax = ax[2])
ax[2].set_title('Hue = day')
sns.despine(left=False)
```

### Distribution Plots

```python
fig, ax = plt.subplots(figsize = (8,5))
sns.histplot(x = 'total_bill', kde = True, data = data, ax=ax)
ax.set_title('Individual Distriubtion Plot with KDE')
sns.despine(left=False)
sns.displot(x = 'total_bill', col = 'day', row = 'sex', binwidth=3, height=3, facet_kws=dict(margin_titles=True), data = data)
sns.despine(left=False)
```

#### Barplot

```python
fig, ax = plt.subplots(1,3, figsize = (18,5))
sns.barplot(x = 'sex', y = 'total_bill', data=data, ax=ax[0])
sns.barplot(x = 'smoker', y = 'total_bill', data=data, ax=ax[1])
sns.barplot(x = 'day', y = 'total_bill', data=data, ax=ax[2])
sns.despine(left=False)
```

#### Violin Plot

```python
fig, ax = plt.subplots(1,2, figsize = (18,6))
sns.violinplot(x = 'day', y = 'total_bill', hue = 'sex', palette='Spectral', split=True, inner="quart", linewidth=1, data=data, ax=ax[0])
ax[0].set_title('hue = sex')
sns.violinplot(x = 'day', y = 'total_bill', hue = 'smoker', palette='Set2', split = True, linewidth=1, data=data, ax=ax[1])
ax[1].set_title('hue = smoker')
sns.despine(left=False)
```

#### Joint Plot

```python
sns.set_style(style='white')
sns.jointplot(x = 'tip', y = 'total_bill', kind = 'scatter', hue = 'smoker', palette = 'Set2', data = data)
sns.jointplot(x = 'tip', y = 'total_bill', kind = 'kde', hue = 'smoker', palette = 'Spectral', data = data)
```

#### Pairplot

```python
sns.set_style(style='ticks')
sns.pairplot(data, hue = 'day', palette = 'Set1')
```

#### Heatmap

```python
fig, ax = plt.subplots(figsize = (9,6))
#`sns.heatmap(data.pivot('tip', 'total_bill', 'size'), vmin = -1, vmax = 1, cmap = 'Spectral', linewidths = 0.5, annot = True, ax=ax[0])
sns.heatmap(data.corr(), vmin = -1, vmax = 1, cmap = 'Spectral', linewidths = 0.5, annot = True, ax=ax)
ax.set_title('Correlation Heatmap')
```

#### LM Plot

```python
sns.lmplot(x = 'tip', y = 'total_bill', col = 'smoker', hue = 'sex', data = data, palette = 'Set1', truncate=False)

sns.lmplot(x = 'tip', y = 'total_bill', col = 'sex', hue = 'time', data = data, palette = 'Set2', truncate=False)
#sns.lmplot(x = 'size', y = 'total_bill', data = data)
```
